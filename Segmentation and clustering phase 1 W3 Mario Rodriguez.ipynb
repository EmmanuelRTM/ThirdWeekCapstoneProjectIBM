```python
import requests
import urllib
from bs4 import BeautifulSoup
source = urllib.request.urlopen('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').read()
soup = BeautifulSoup(source, 'html.parser')
# read wikipedia Link
```

```python
My_table = soup.find('table',{'class':'wikitable sortable'})
#My_table
#sparse referenced table
```

```python
import re
#import library to use regular expressions
```

```python
table_2=[]
x_tr=My_table.find_all('tr')
for i in x_tr:
    try:
        for j in i.find_all('td'):
            try:  
                x_text=re.sub(r"(?<=[a-z])\r?\n","", j.get_text())
                table_2.append(x_text)
            except:
                next        
    except:
        next
#create a list of all text components of the table
```

```python
table_k=[]
table_f=[]
count=0
for k in table_2:
    try:
        count=count+1
        if count==4:
            table_f.append(table_k)
            table_k=[]
            count=1
            #k.replaceAll('\n','')
            table_k.append(k)
        else:
            #k.replaceAll('\n','')
            table_k.append(k)
    except:
        next
# put each line of the wikipedia table in list of trios and then put them inside a list
```

```python
import pandas as pd
df=pd.DataFrame(table_f,columns=['PostalCode','Borough','Neighborhood'])
df.head()
##transform previous list into a data frame to make it into a table.
```

```python
#erase all lines in Borough column within "Not assigned" text.
df1=df
df1=df1[df1.Borough != 'Not assigned']
df1.head()

#Find each line of Neighborhood column that has the text "Not assigned" and then copy the text from Borough column with the respective row.
index_list=df1.index[df1.Neighborhood=='Not assigned'].tolist()
df1.loc[min(index_list):max(index_list),'Neighborhood'] = df1.loc[min(index_list):max(index_list),'Borough'].to_numpy()
#

df1.head(10)
```

```python
#filter unique text in postalcode column and join text with commas from neighborhood column into a single row per unique PostalCode value.
df2=df1.groupby("PostalCode").agg(lambda x:', '.join(set(x)))
# put PostalCode column as a normal column by reseting the index
df2.reset_index(level=0, inplace=True)
df2.head(15)
#print table
```
