```python
import requests
import urllib
from bs4 import BeautifulSoup
source = urllib.request.urlopen('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').read()
soup = BeautifulSoup(source, 'html.parser')
# read wikipedia Link
```

```python
My_table = soup.find('table',{'class':'wikitable sortable'})
#My_table
#sparse referenced table
```

```python
import re
#import library to use regular expressions
```

```python
table_2=[]
x_tr=My_table.find_all('tr')
for i in x_tr:
    try:
        for j in i.find_all('td'):
            try:  
                x_text=re.sub(r"(?<=[a-z])\r?\n","", j.get_text())
                table_2.append(x_text)
            except:
                next        
    except:
        next
#create a list of all text components of the table
```

```python
table_k=[]
table_f=[]
count=0
for k in table_2:
    try:
        count=count+1
        if count==4:
            table_f.append(table_k)
            table_k=[]
            count=1
            #k.replaceAll('\n','')
            table_k.append(k)
        else:
            #k.replaceAll('\n','')
            table_k.append(k)
    except:
        next
# put each line of the wikipedia table in list of trios and then put them inside a list
```

```python
import pandas as pd
df=pd.DataFrame(table_f,columns=['PostalCode','Borough','Neighborhood'])
df.head()
##transform previous list into a data frame to make it into a table.
```

```python
#erase all lines in Borough column within "Not assigned" text.
df1=df
df1=df1[df1.Borough != 'Not assigned']
df1.head()

#Find each line of Neighborhood column that has the text "Not assigned" and then copy the text from Borough column with the respective row.
index_list=df1.index[df1.Neighborhood=='Not assigned'].tolist()
df1.loc[min(index_list):max(index_list),'Neighborhood'] = df1.loc[min(index_list):max(index_list),'Borough'].to_numpy()
#

df1.head(10)
```

```python
# function that extracts the category of the venue that is required
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

```python
venues = results['response']['groups'][0]['items']
    
nearby_venues = json_normalize(venues) # flatten JSON

# filter columns
filtered_columns = ['venue.name', 'venue.categories', 'venue.location.lat', 'venue.location.lng']
nearby_venues =nearby_venues.loc[:, filtered_columns]

# filter the category for each row
nearby_venues['venue.categories'] = nearby_venues.apply(get_category_type, axis=1)

# clean columns
nearby_venues.columns = [col.split(".")[-1] for col in nearby_venues.columns]

nearby_venues.head()
```

```python
#automiatize the previous step in order to do it with all obtained venues
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
#apply previous function
Toronto_venues = getNearbyVenues(names=Toronto_df['Neighborhood'],
                                   latitudes=Toronto_df['Latitude'],
                                   longitudes=Toronto_df['Longitude']
                                  )
```

```python
#Analize neighborhood and use get_dummies function to get 0 1 according to each variable in order to put the Dataframe into a ML model.

Toronto_offwords = pd.get_dummies(Toronto_venues[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
Toronto_offwords['Neighborhood'] = Toronto_venues['Neighborhood'] 
Toronto_offwords.head()
# move neighborhood column to the first column

#Create a list of the name of the columns, then put Neighborhood column as the first one.
cols = Toronto_offwords.columns.tolist()
n = int(cols.index('Neighborhood'))
cols = [cols[n]] + cols[:n] + cols[n+1:]
Toronto_offwords = Toronto_offwords[cols]

Toronto_offwords.head()
```

```python
Toronto_offwords.shape
```

```python
## group rows by neighborhood and by taking the mean of the frequency of occurrence of each category
Toronto_grouped = Toronto_offwords.groupby('Neighborhood').mean().reset_index()
Toronto_grouped.head()
```

```python
Toronto_grouped.shape
```

```python
#print each neighborhood along with the top 5 most common venues
num_top_venues = 5

for hood in Toronto_grouped['Neighborhood']:
    print("----"+hood+"----")
    temp = Toronto_grouped[Toronto_grouped['Neighborhood'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

```python
#Define a function to sort venues in descending order
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
#create a new dataframe and display the top 10 venues for each neighborhood.
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhood'] = Toronto_grouped['Neighborhood']

for ind in np.arange(Toronto_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(Toronto_grouped.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted.head()
```

```python
##Run k-means to cluster the neighborhood into 5 clusters.

# set number of clusters
kclusters = 5

Toronto_grouped_clustering = Toronto_grouped.drop('Neighborhood', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(Toronto_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

```python
#create a new dataframe that includes the cluster as well as the top 10 venues for each neighborhood.

# add clustering labels
neighborhoods_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)
Toronto_data1=Toronto_df
Toronto_merged = Toronto_data1

# merge toronto_grouped with toronto_data1 to add latitude/longitude for each neighborhood
Toronto_merged = Toronto_merged.join(neighborhoods_venues_sorted.set_index('Neighborhood'), on='Neighborhood')

Toronto_merged.head() # check the last columns!
```

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(Toronto_merged['Latitude'], Toronto_merged['Longitude'], Toronto_merged['Neighborhood'], Toronto_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
